## FUNCTIONS USED IN THE CODE

```python
def to_json(file_path, dictionary):
    '''
  input: file_path string, python dictionary
  output: none
  action:  writes a file with content of the dictionary
    '''
    with open(file_path, 'w') as json_file:
        json.dump(dictionary, json_file)

def to_dict(file_path):
  '''
  input: file_path string
  output: python dictionary
  action: loads and returns a dictionay of the given json file path
  '''
  return json.load(open(file_path, 'r'))


#FUNCTION USED TO CHANGE THE SCHEMA OF A PARTICULAR FIELD (STRING-> FLOAT-> INTEGER)
def convert_integer(x):
    try:
        value = int(float(x))
    except ValueError:
         value = 0
    return value


#FUNCTION USED TO MERGE MULTIPLE DOMAINS
def merge_domain(x,vocabulary):
    try:
        out = vocabulary[x]
    except KeyError:
        print('Palakshee you missed this mapping in vocabulary ->', x)
    return out
```

## DATA CLEANING

```python
#READING OUR MASTER FILE USING PANDAS TO DO FURTHER DATA CLEANING 
master =  pd.read_csv('./csv/master.csv')

#DROPING THE TIMESTAMP FIELD
master.drop(['Timestamp'],axis=1 ,inplace = True)
```

```python
#CHANGING THE SCHEMA OF AGE ATTRIBUTE USING PERVIOUSLY DEFINED FUNCTION
master['Age'] = master['Age'].apply(convert_integer)


#FILLING THE NULL(NAN) VALUES OF OCCUPATION, STATE, GENDER ATTRIBUTE WITH "UNKNOWN"
master['Occupation'] = master['Occupation'].fillna('unknown')
master['State'] = master['State'].fillna('unknown')
master['Gender'] = master['Gender'].fillna('unknown')


#CLEANING OF Occupation ATTRIBUTE
#USING merge_domain FUNCTION
#MAPPING MULTIPLE  SIMILAR DOMAINS IN A SINGLE DOMAIN for example Student,Schoolgirl,studying ...etc IN student) 
occupation_vocabulary = {'Student':'student','Schoolgirl':'student','Students':'student', 'Student ':'student','Studying':'student','student ':'student', 'student':'student','12th ':'student','Broke':'student','School':'student','School ':'student','Tera baap':'student','homosapien':'student','ningen':'student','Homo sapien':'student', 'Business man':'business_man','Business man ':'business_man','Business man and student':'business_man','Clown':'service','Service':'service','Chef':'service','Musician ':'service','Working professional':'service','Doctor':'service','Software developer':'service','Employer ':'service','Engineer':'service','Fashion designer':'service','Design engineer':'service','Govt':'service','Lawyer':'service','IT PROFESIONAL ':'service','Job':'service','Manager':'service','Optician':'service','Musician':'service','Professional':'service','Professional ':'service','Software':'service','Academic':'service','Teacher':'teacher','Teachee':'teacher','Teaching':'teacher', 'homemaker':'homemaker','Housemaker':'homemaker','Housemaker ':'homemaker','unknown':'unknown','-':'unknown','Sameer Saini':'unknown','No':'unknown'  }
master['Occupation'] = master['Occupation'].apply(lambda x: merge_domain(x, occupation_vocabulary))
master
```

```python
master['State']= master['State'].str.lower() 
master['State']= master['State'].str.strip(' ') 
master['State']= master['State'].str.split().str.get(0)
master['State']= master['State'].str.strip(',') 
```

```python
#CLEANING OF State ATTRIBUTE
#USING merge_domain FUNCTION
#MAPPING MULTIPLE  SIMILAR DOMAINS IN A SINGLE DOMAIN 9for example h.p,himachal,himachal pradesh ...etc IN Himachal) 


state_vocabulary = {'Delhi':'Delhi','Panjab':'Punjab','Punjan':'Punjab','PUNJAB':'Punjab','Manimajra Chandigarh':'Punjab','panchkula ':'Punjab','Punjab ':'Punjab','Punjab':'Punjab','manimajra':'Punjab','punjab':'Punjab','panchkula':'Punjab','mohali(punjab)':'Punjab','mohali':'Punjab','panjab':'Punjab','punjan':'Punjab','haryana':'Haryana','haryaja':'Haryana','Haryana':'Haryana','Haryana ':'Haryana','Uttar Pradesh':'Uttar_Pradesh','uttar':'Uttar_Pradesh','uttar pradesh':'Uttar_Pradesh','up':'Uttar_Pradesh','Uttar pardesh':'Uttar_Pradesh','Uttar Pradesh ':'Uttar_Pradesh','telangana':'Telangana','Telangana':'Telangana', 'CHANDIGARH (UT)':'Chandigarh','Chandigarh':'Chandigarh','Chandigarh ':'Chandigarh','Chandigarh UT':'Chandigarh','chandigarh':'Chandigarh', 'ut':'Chandigarh','Ut Chandigarh ':'Chandigarh','maharashtra':'Maharashtra','Maharashtra':'Maharashtra','Maharashtra ':'Maharashtra','Himachal ':'Himachal','Himachal':'Himachal','himachal':'Himachal','h.p':'Himachal','H.p':'Himachal','Himachal pradesh ':'Himachal','Himachal Pradesh':'Himachal','delhi':'Delhi','uttarakhand':'Uttarakhand','Uttarakhand':'Uttarakhand','Uttarakhand ':'Uttarakhand','Karnataka':'Karnataka','karnataka':'Karnataka','Karnataka ':'Karnataka','assam':'Assam','assamm':'Assam','west':'West_Bengal','bihar':'Bihar','gujarat':'Gujarat','bengal':'West_Bengal','Bengal':'West_Bengal','West Bengal':'West_Bengal', 'ontario':'International_State','usa':'International_State','california':'International_State','good':'unknown','depression':'unknown','not':'unknown','happy':'unknown','worse':'unknown','your':'unknown','healthy':'unknown','better':'unknown','solid':'unknown','-':'unknown','unknown':'unknown'}
master['State'] = master['State'].apply(lambda x: merge_domain(x, state_vocabulary))
master
```

```python
#FILLING THE NULL VALUES WITH 0 IN THE IMAGES FIELDS
#REPLACING LIKE WITH 1 AND DISLIKE WITH -1 IN THE IMAGE FIELDS

master= master.fillna(0)
master= master.replace('Like',1)
master= master.replace('Dislike',-1)
master
```

```python
#REPLACING THE UNKNOWN VALUES OF STATE AND GENDER FIELDS WITH THE MAXIMUM OCCURING VALUES OF THE RESPECTIVE FIELDS 
#REPLACING THE UNKNOWN OF AGE FIELD WITH THE MEAN OF THE AGE FIELD
max_state=master['State']. value_counts(). idxmax()
max_gender=master['Gender']. value_counts(). idxmax()
mean_value = master['Age'].mean()
age_mean = int(mean_value)
master['State']= master['State'].replace('unknown',max_state)
master['Gender']= master['Gender'].replace('unknown',max_gender)
master['Age']= master['Age'].replace(0,age_mean)
master
```
